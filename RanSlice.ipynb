```
# Import Statements
from graph_tool.all import *
import matplotlib.pyplot as pypt
import math
```

```
# Declaring Global Variables

ranSlices = []
vnfFunctions = []
resList = [3, 2, 1]
vnfConnections = [(0, 1), (0, 2), (1, 2)]
vnfEdges = []
```

```
# Function defined to create an RAN Slice

def createRANSlices():

    # Creating the RAN Slices

    for idx in range(2):
        ranSlices.append(Graph(directed=False))
```

```
# Setting Properties for the RAN Slice

def setRANSliceProperties():

    # Creating Properties for the RAN Slices
    # -------------------------------------- 

    for counterVar in range(len(ranSlices)):

      # Giving the VNF Functions a Name Property
      nameProp = ranSlices[counterVar].new_vertex_property("string")
      ranSlices[counterVar].vertex_properties.vertexName = nameProp

      # Giving the VNF Functions a Resource Capacity Property
      resourceCapacityProp = ranSlices[counterVar].new_vertex_property("int")
      ranSlices[counterVar].vertex_properties.resourceCapacity = resourceCapacityProp

      # Giving the VNF Functions a Resource Property
      resourceProp = ranSlices[counterVar].new_vertex_property("int")
      ranSlices[counterVar].vertex_properties.resources = resourceProp

      # Giving the VNF Functions a Bandwidth Property
      capacityProp = ranSlices[counterVar].new_edge_property("int")
      ranSlices[counterVar].edge_properties.bandwidth = capacityProp

      # Giving the Substrate Vertices a Resource Capacity Property
      binaryMappingVar = ranSlices[counterVar].new_vertex_property("int")
      ranSlices[counterVar].vertex_properties.binaryMappingVar = binaryMappingVar  
```

```
# Function to create VNF Functions for the respective RAN Slices

def createVNFFunctions():
    
  # Creating the VNF Functions

  for counterVar in range(len(ranSlices)):
    for idx in range(3):
      ranVertex = ranSlices[counterVar].add_vertex()
```

```
# Function to Set the Properties for the VNF Functions

def setVNFFunctionProperties():

  # Setting the names for the VNF Functions

  for counterVar in range(len(ranSlices)):  
    for idx in range(3):
      ranSlices[counterVar].vertex_properties.vertexName[vnfFunctions[idx]] = "S" + str(idx)
      # Setting the resource capacity for the VNF Functions
      ranSlices[counterVar].vertex_properties.resourceCapacity[vnfFunctions[idx]] = -1
      # Setting the resource value for the VNF Functions
      ranSlices[counterVar].vertex_properties.resources[vnfFunctions[idx]] = resList[idx]
      # Setting the Binary Mapping Variable for the VNF Functions
      ranSlices[counterVar].vertex_properties.binaryMappingVar[vnfFunctions[idx]] = 0
      
```

```
# Function to create Connections within the VNF Functions

def createVNFConnections():
  # Creating the connections between VNF Functions

  for counterVar in range(len(ranSlices)):
    for vnf_connection in vnfConnections:
      ranSlicesEdges = ranSlices[counterVar].add_edge(vnf_connection[0], vnf_connection[1])
      vnfEdges.append(ranSlicesEdges) 

  # Setting the Bandwidth Property on the Connections between the Substrate Networks
  # !! Hard Coded Value Included within this Loop !!

  for counterVar in range(len(ranSlices)): 
    for edgeConnection in vnfEdges:
      ranSlices[counterVar].edge_properties.bandwidth[edgeConnection] = 3
```

```
def initRANSlices():
  createRANSlices()
  setRANSliceProperties()
  createVNFFunctions()
  setVNFFunctionProperties()
  createVNFConnections()
```
